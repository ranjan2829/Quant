```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
import yfinance as yf
import os
```

```python
plt.style.use('ggplot')
sentiment=pd.read_csv('sentiment_data.csv')
```

```python
sentiment['date']=pd.to_datetime(sentiment['date'])
sentiment

```

```python
sentiment=sentiment.set_index(['date','symbol'])
```

```python
sentiment
```

```python
sentiment['engagement_ratio']=sentiment['twitterComments']/sentiment['twitterLikes']
sentiment=sentiment[(sentiment['twitterLikes']>20)&(sentiment['twitterComments']>10)]
```

```python
sentiment
```

```python
aggragate=(sentiment.reset_index('symbol').groupby([pd.Grouper(freq='M'),'symbol'])
           [['engagement_ratio']].mean())
```

```python
aggragate['rank']=(aggragate.groupby(level=0)['engagement_ratio']
                   .transform(lambda x: x.rank(ascending=False)))
```

```python
aggragate
```

```python
filter =aggragate[aggragate['rank']<6].copy()
filter=filter.reset_index(level=1)
filter.index=filter.index+pd.DateOffset(1)
filter=filter.reset_index().set_index(['date','symbol'])
filter.head(20)
```

```python
dates=filter.index.get_level_values('date').unique().tolist()
fixed={}
for d in dates:
  fixed[d.strftime('%Y-%m-%d')]=filter.xs(d,level=0).index.tolist()
fixed
```

```python

stocks_list = sentiment.index.get_level_values('symbol').unique().tolist()

# Remove the problematic stock 'ATVI' from the list
stocks_list.remove('ATVI')

# Download prices for the remaining stocks
prices_data = yf.download(tickers=stocks_list,
                          start='2021-01-01',
                          end='2023-03-01')
```

```python
import numpy as np
import pandas as pd

# Assuming you already have prices_df and fixed defined

returns_df = np.log(prices_df['Adj Close']).diff().dropna()

portfolio_df = pd.DataFrame()

for start_date in fixed.keys():
    end_date = (pd.to_datetime(start_date) + pd.offsets.MonthEnd()).strftime('%Y-%m-%d')
    cols = fixed[start_date]

    # Print for debugging
    print(f"Processing date range: {start_date} to {end_date}")
    print(f"Selected columns: {cols}")

    selected_data = returns_df[start_date:end_date][cols]

    # Check for NaN values in the selected data
    if selected_data.isnull().values.any():
        print("Warning: NaN values found in the selected data. Check for missing or incomplete data.")

    temp_df = selected_data.mean(axis=1).to_frame('portfolio_return')

    # Print intermediate results
    print(f"Portfolio returns for date range {start_date} to {end_date}:")
    print(temp_df)

    portfolio_df = pd.concat([portfolio_df, temp_df], axis=0)

# Print final result
print("Final portfolio returns:")
print(portfolio_df)
```
