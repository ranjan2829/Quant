The geometric Brownian motion (GBM) model is a mathematical model used to describe the stochastic process of asset prices over time. It is a continuous-time stochastic process widely employed in financial modeling, particularly for modeling the behavior of stock prices.

In the context of finance, the GBM model is commonly used to model the dynamics of stock prices, although it can be applied to other financial instruments as well. The model assumes that the logarithm of the asset price follows a Brownian motion with drift and volatility.

The stochastic differential equation (SDE) governing the GBM process is:

\[ dS_t = \mu S_t dt + \sigma S_t dZ_t \]

where:
- \( S_t \) is the price of the asset at time \( t \).
- \( \mu \) is the drift coefficient, representing the expected return of the asset per unit of time.
- \( \sigma \) is the volatility coefficient, representing the standard deviation of the asset's returns per unit of time.
- \( Z_t \) is a standard Brownian motion, representing random market movements.

The GBM model is widely used in option pricing, risk management, and portfolio optimization due to its simplicity and analytical tractability. It serves as a foundation for more complex models and techniques used in quantitative finance.

```python
import math
import numpy as np
import pandas as pd
import scipy.stats as scs
import statsmodels.api as sm
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['font.family'] = 'serif'
```

```python
def dN(x,mu,sigma):
  z=(x-mu)/sigma
  pdf=np.exp(-0.5*z**2)/math.sqrt(2*math.pi*sigma**2)
  return pdf
```

```python
def simulate_gbm():
    # model parameters
    S0 = 100.0  # initial index level
    T = 10.0  # time horizon
    r = 0.05  # risk-less short rate
    vol = 0.2  # instantaneous volatility

    # simulation parameters
    np.random.seed(250000)
    gbm_dates = pd.DatetimeIndex(start='30-09-2004',
                                 end='30-09-2014',
                                 freq='B')
    M = len(gbm_dates)  # time steps
    I = 1  # index level paths
    dt = 1 / 252.  # fixed for simplicity
    df = math.exp(-r * dt)  # discount factor

    # stock price paths
    rand = np.random.standard_normal((M, I))  # random numbers
    S = np.zeros_like(rand)  # stock matrix
    S[0] = S0  # initial values
    for t in range(1, M):  # stock price paths
        S[t] = S[t - 1] * np.exp((r - vol ** 2 / 2) * dt +
                                 vol * rand[t] * math.sqrt(dt))

    gbm = pd.DataFrame(S[:, 0], index=gbm_dates, columns=['index'])
    gbm['returns'] = np.log(gbm['index'] / gbm['index'].shift(1))

    # Realized Volatility (eg. as defined for variance swaps)
    gbm['rea_var'] = 252 * np.cumsum(gbm['returns'] ** 2) / np.arange(len(gbm))
    gbm['rea_vol'] = np.sqrt(gbm['rea_var'])
    gbm = gbm.dropna()
    return gbm
```
